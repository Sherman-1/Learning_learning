```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=0)
y = y.reshape(y.shape[0],1)


def initialisation(X):

    W = np.random.rand(X.shape[1],1)
    b = np.random.rand(1)

    return W,b

def model(X,W,b):

    Z = X.dot(W) + b
    A = 1 / (1 + np.exp(-Z))
    return A

def log_loss(A,y):

    return 1/len(y) * np.sum(y * np.log(A) + (1-y) * np.log(1-A))

def gradient_descent(A,X,y):

    dW = 1 / len(y) * np.dot(X.T, (A - y))
    db = 1 / len(y) * np.sum(A - y)

    return dW,db

def update_parameters(W,b,dW,db,learning_rate):

    W = W - learning_rate * dW
    b = b - learning_rate * db

    return W,b

```

```python
class Neuron():

    def __init__(self, X):
        self.W, self.b = initialisation(X)

    def train(self, X, y, epochs = 100):

        import matplotlib.pyplot as plt
        losses = []
        for i in range(epochs):

            A = model(X,self.W,self.b)
            loss = log_loss(A,y)
            losses.append(loss)
            dW, db = gradient_descent(A,X,y)
            self.W, self.b = update_parameters(self.W,self.b,dW,db,0.01)

            if i % 100 == 0:
                print(f"Loss at epoch {i} is {loss}")

        plt.plot(losses)
    def predict(self, X):

        return model(X,self.W,self.b)
    
from sklearn import datasets
diabetes = datasets.load_diabetes()

X = diabetes.data
y = diabetes.target.reshape(-1,1)

neuron = Neuron(X)
neuron.train(X,y,10000)
```
